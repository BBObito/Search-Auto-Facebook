Selenium không có một lệnh trực tiếp như "click new tab", nhưng nó hỗ trợ mở liên kết trong một tab mới hoặc chuyển đổi giữa các tab. Điều này có thể thực hiện bằng cách sử dụng tổ hợp phím hoặc điều hướng đến URL cụ thể trong tab mới.

### Cách mở tab mới trong Selenium:

1. **Dùng tổ hợp phím**: Selenium hỗ trợ thực hiện các tổ hợp phím để mở tab mới. Ví dụ:
   ```python
   from selenium import webdriver
   from selenium.webdriver.common.keys import Keys

   # Khởi tạo trình duyệt
   driver = webdriver.Chrome()

   # Mở tab mới
   driver.get("https://example.com")
   body = driver.find_element("tag name", "body")
   body.send_keys(Keys.CONTROL + 't')  # Với MacOS, thay `CONTROL` bằng `COMMAND`

   # Chuyển đến tab mới
   driver.switch_to.window(driver.window_handles[-1])
   driver.get("https://newtab.com")
   ```

2. **Sử dụng JavaScript để mở tab mới**:
   Bạn có thể dùng JavaScript để mở tab mới thông qua `execute_script`:
   ```python
   driver.execute_script("window.open('https://newtab.com', '_blank');")
   driver.switch_to.window(driver.window_handles[-1])  # Chuyển sang tab mới
   ```

3. **Click mở tab mới từ liên kết**:
   Nếu bạn muốn nhấp vào một liên kết mà mặc định mở ra tab mới, bạn không cần làm gì đặc biệt, chỉ cần nhấp vào liên kết đó:
   ```python
   link = driver.find_element("link text", "Open new tab")
   link.click()

   # Sau khi tab mới mở, chuyển sang tab đó
   driver.switch_to.window(driver.window_handles[-1])
   ```

4. **Quản lý nhiều tab**:
   Khi làm việc với nhiều tab, bạn cần dùng `driver.window_handles` để quản lý các cửa sổ hoặc tab:
   ```python
   for handle in driver.window_handles:
       driver.switch_to.window(handle)
       print(driver.title)  # Xem tiêu đề của từng tab
   ```

### Kết luận
Selenium không có hàm `click new tab` cụ thể, nhưng bạn có thể:
- Mở tab mới bằng tổ hợp phím.
- Sử dụng JavaScript để mở tab mới.
- Xử lý liên kết tự động mở tab mới.
