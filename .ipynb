Selenium khng c mt lnh trc tip nh "click new tab", nhng n h tr m lin kt trong mt tab mi hoc chuyn i gia cc tab. iu ny c th thc hin bng cch s dng t hp phm hoc iu hng n URL c th trong tab mi.

### Cch m tab mi trong Selenium:

1. **Dng t hp phm**: Selenium h tr thc hin cc t hp phm  m tab mi. V d:
   ```python
   from selenium import webdriver
   from selenium.webdriver.common.keys import Keys

   # Khi to trnh duyt
   driver = webdriver.Chrome()

   # M tab mi
   driver.get("https://example.com")
   body = driver.find_element("tag name", "body")
   body.send_keys(Keys.CONTROL + 't')  # Vi MacOS, thay `CONTROL` bng `COMMAND`

   # Chuyn n tab mi
   driver.switch_to.window(driver.window_handles[-1])
   driver.get("https://newtab.com")
   ```

2. **S dng JavaScript  m tab mi**:
   Bn c th dng JavaScript  m tab mi thng qua `execute_script`:
   ```python
   driver.execute_script("window.open('https://newtab.com', '_blank');")
   driver.switch_to.window(driver.window_handles[-1])  # Chuyn sang tab mi
   ```

3. **Click m tab mi t lin kt**:
   Nu bn mun nhp vo mt lin kt m mc nh m ra tab mi, bn khng cn lm g c bit, ch cn nhp vo lin kt :
   ```python
   link = driver.find_element("link text", "Open new tab")
   link.click()

   # Sau khi tab mi m, chuyn sang tab 
   driver.switch_to.window(driver.window_handles[-1])
   ```

4. **Qun l nhiu tab**:
   Khi lm vic vi nhiu tab, bn cn dng `driver.window_handles`  qun l cc ca s hoc tab:
   ```python
   for handle in driver.window_handles:
       driver.switch_to.window(handle)
       print(driver.title)  # Xem tiu  ca tng tab
   ```

### Kt lun
Selenium khng c hm `click new tab` c th, nhng bn c th:
- M tab mi bng t hp phm.
- S dng JavaScript  m tab mi.
- X l lin kt t ng m tab mi.
